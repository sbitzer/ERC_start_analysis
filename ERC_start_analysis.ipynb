```python
import pandas as pd
```

```python
grantees = pd.read_csv('ERC_grantees_career_state.csv', sep=';', 
                       skipinitialspace=True)
```

```python
grantees
```

```python
# group leaders are postdocs which are no group leaders
ispostdoc = grantees['position'].str.contains(r'[pP]ostdoc')
isgrplead = grantees['position'].str.contains(r'[gG]roup leader')
ispostdoc = ispostdoc & ~isgrplead
print('number of postdocs: %d' % ispostdoc.sum())
```

```python
# separate Juniorprofessors and full professors
isprof = grantees['position'].str.contains(r'[pP]rofessor')
isjunprof = grantees['position'].str.contains(r'[jJ]uniorprofessor')
isfullprof = isprof & ~isjunprof
print('number of full professors: %d' % isfullprof.sum())
```

```python
# full professors are no junior group leaders, but Juniorprofs are
isjuniorgrplead = (isgrplead & ~isfullprof) | isjunprof
print('number of junior research group leaders: %d' % isjuniorgrplead.sum())
```

```python
# account for missing entries by summing only over identified positions
print('percentage postdocs: %4.1f%%' % (ispostdoc.sum() / float(
    ispostdoc.sum() + isfullprof.sum() + isjuniorgrplead.sum()) * 100))
```
